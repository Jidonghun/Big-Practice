### 자료형

#### 1. 문자
 * 기본 자료형
     - 문자열: str ...
     ```python
        print("Hi")
        print("내 꿈은 '멋진 집'을 짓는 것이다")
        print("세종"+"대왕")
     ```
     - 숫자: int, float ...
     - 불(boolean): True, False ...
    
 * 이스케이프 문자
     - \n: 줄바꿈
     - \t: 탭

 * 인덱스
     - 첫 글자는 인덱스 0
     - 마지막 글자는 인덱스 -1
     ```python
        print("가나다라마바사"[2:5])
     ```
     - 위 코드 경우 [2:5] -> 2, 3, 4

 * 문자열 길이 구하기
     - len()
     
#### 2. 숫자
 * 숫자 연산자
     - 정수 나누기: //
     - 나머지 연산자: %
     - 제곱 연산자: **
     
#### 3. 변수
 * 복합 대입 연산자
     - +=: 숫자 덧셈 후 대입
     - -=: 숫자 뺄셈 후 대입
     - *=: 숫자 곱셈 후 대입
     - /=: 숫자 나눗셈 후 대입
     - %=: 숫자의 나머지를 구한 후 대입
     - **=: 숫자 제곱 후 대입
     
 * 문자 -> 숫자 변환
     - int(): 정수
     - float(): 실수

 * 숫자 -> 문자 변환
     - str()

#### 4. 숫자와 문자열의 다양한 기능
 * 문자열의 format() 함수
     ```python
        print('{}의 높이는 {}[m]입니다.'.format('설악산', '1188'))
     ```

 * 정수 출력의 형태
     ```python
        print('{:d}'.format(34))
        print('{:f}'.format(3.14))
        print('{:0.2f}'.format(3.1415))
     ```

 * 대소문자 바꾸기
     ```python
        name = 'donghun ji'
        print(name.upper())
        print(name.lower())
     ```

 * 문자열 양옆의 공백 제거
     ```python
        price = '    125,000원'
        print(price)
        print(price.strip())
     ```

 * 문자열 구성 파악
    - isalpha(): 문자열이 알파벳으로만 구성되어 있는지
    - isidentifier(): 문자열이 숫자로만 구성되어 있는지
    - isdecimal(): 문자열이 정수형태인지
    - isdigit(): 문자열이 숫자를 나타내는 문자로만 구성되어 있는지, 다른 말로 문자가 하나라도 있으면 False

 * 문자열 찾기
    - find(): 문자열 왼쪽부터 검색
    - rfind(): 문자열 오른쪽부터 검색

 * 문자열과 in 연산자
     ```python
        '종이' in '종이 땡땡땡'
        True
     ```  

 * 문자열 자르기
    - split()
     ```python
        hobby = '드라이브, 낚시, 등산'
        print(hobby.split(','))
     ```

 * 문자 대체
    - replace()
     ```python
        name = 'Ji, dong-hun'
        name = name.replace('-', '')
        print(name)
     ```  

 * f-문자열
     ```python
        name = 'Ji donghun'
        hobby = 'drive'
        address = 'Pohang'

        print(f'{name}의 취미는 {hobby}이고 사는 곳은 {address}입니다.')
     ```
